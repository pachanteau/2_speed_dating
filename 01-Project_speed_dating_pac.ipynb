```python
# filter initial dataset only on people who meet each other before & took decison to meet people again ...

df2 = df[(df["dec"]==1) & (df["met"]==2)]
```

```python
# Check df filtered

df2
```

```python
# filtering on most relevant keys for all waves mixed ...


filter = (df["dec"]==1) & (df["met"]==2)

df3 = df.loc[filter,['dec','attr','sinc','intel','fun','amb','shar','like','prob']]
df3
```

```python
df_attributes_score = df3.groupby(["dec"]).sum().transpose().sort_values(by=1,ascending=False)
```

```python
df_attributes_score = df_attributes_score.reset_index()
```

```python
df_attributes_score.columns= ['attributes_codes','ratings_sum']
df_attributes_score
```

```python
# df_wave_genre['gender_labels'] = df_wave_genre['gender'].apply(lambda x : 'Female' if x==0 else 'Male')

df_attributes_score['attributes_labels'] = df_attributes_score['attributes_codes'].apply(lambda x : 
                                                                                         'Attractive' if x=='attr'
                                                                                         else 'Sincere' if x=='sinc'
                                                                                         else 'Intelligent' if x=='intel'
                                                                                         else 'Fun' if x=='fun'
                                                                                         else 'Ambitious' if x=='amb'
                                                                                         else 'Shared Interests/Hobbies' if x=='shar'
                                                                                        else 'like' if x=='like'
                                                                                        else 'probability')
```

```python
df_attributes_score
```

```python
fig = plt.figure(figsize=(15,5))
ax = fig.add_axes([0,0,1,1])
co1 = 'lightblue'
ec = ['green', 'green', 'green', 'blue', 'white', 'white', 'blue','white']
x = df_attributes_score['attributes_labels']
y = df_attributes_score['ratings_sum']
ax.bar(x,y,color=[co1], edgecolor = ec)

plt.title('what are the attribute attractivness ?', fontsize= 30)
plt.xlabel('attributes', fontsize=20)
plt.ylabel('scoring', fontsize=20)

plt.show()
```
